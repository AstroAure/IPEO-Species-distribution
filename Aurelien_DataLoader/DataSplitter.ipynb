```python
# Find project root

from pathlib import Path
try:
    root = Path(__file__).resolve().parent
except NameError:
    root = Path.cwd()  # fallback for Jupyter notebooks

while root.parent != root:
    if any((root / marker).exists() for marker in ["README.md"]):
        break
    root = root.parent

# Fallback in case nothing found
if not any((root / marker).exists() for marker in ["README.md"]):
    print("Could not locate project root — defaulting to current working directory")
    root = Path.cwd()

root = str(root)
print(f"Root folder detected at: {root}")
```

```python
import numpy as np
import pandas as pd
```

```python
env_variables = pd.read_csv(f"{root}/data/env_variables_training.csv")
landsat_timeseries = pd.read_csv(f"{root}/data/landsat_timeseries_training.csv")
satellite_patches = np.load(f"{root}/data/satellite_patches_training.npy")
species_labels = np.load(f"{root}/data/species_data_training.npy")
```

```python
val_idx = np.random.choice(len(env_variables), size=int(0.2*len(env_variables)), replace=False)
```

```python
len(val_idx)
```

```python
env_variables_val = env_variables.iloc[val_idx]
landsat_timeseries_val = landsat_timeseries.iloc[val_idx]
satellite_patches_val = satellite_patches[val_idx]
species_labels_val = species_labels[val_idx]

env_variables_train = env_variables.drop(val_idx)
landsat_timeseries_train = landsat_timeseries.drop(val_idx)
satellite_patches_train = np.delete(satellite_patches, val_idx, axis=0)
species_labels_train = np.delete(species_labels, val_idx, axis=0)
```

```python
env_variables_train.to_csv(f"{root}/data/env_variables_train.csv", index=False)
landsat_timeseries_train.to_csv(f"{root}/data/landsat_timeseries_train.csv", index=False)
np.save(f"{root}/data/satellite_patches_train.npy", satellite_patches_train)
np.save(f"{root}/data/species_labels_train.npy", species_labels_train)

env_variables_val.to_csv(f"{root}/data/env_variables_val.csv", index=False)
landsat_timeseries_val.to_csv(f"{root}/data/landsat_timeseries_val.csv", index=False)
np.save(f"{root}/data/satellite_patches_val.npy", satellite_patches_val)
np.save(f"{root}/data/species_labels_val.npy", species_labels_val)
```
